```python
#Question 2

#Write a program which will find all such numbers which are divisible by 7 but are not a multiple
#of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a
#comma-separated sequence on a single line.

#Solution:
#Algorithm:

# 1. Get both Numbers from the user in case of dynamic input and store it in 2 variables.
# 2. Use Range inside for loop between Both the numbers.
# 3. Loop the numbers inside If statement and use modulus operator to find the number is divisible by 7 but are not a multiple 5.
# 4. Append the number in an array.
# 5. Print the result.

# Program
# Getting input Maximum and Minimum From the user

Minimum = int(input("Enter your Minumum Number : "))
Maximum = int(input("Enter your Maximum Number : "))
Num=[]

# Checking Minimum number is greater than Maximum Number

if (Minimum > Maximum):
    print("Min cannot be greater than max")

# Looping between Minimum and Maximum Number and store the next number in Result_Num variable.
    
for Result_Num in range(Minimum, Maximum):
    #Checking the number is divisible by 7 and not divisible by 5
    if (Result_Num%7==0) and (Result_Num%5!=0):
    #Appending the result in Num array
        Num.append(str(Result_Num))
#Print the array
print('-' *125)
print("\nNumbers which are divisible by 7 but are not a multiple of 5 : \n")
print(Num,'\n')
print('-' *125)
```

```python
#Question 3

#Write a Python program to accept the user's first and last name and then getting them printed in the the 
#reverse order with a space between first name and last name.

#Solution:

# 1. Get both First Name and Last Name from the user in case of dynamic input and store it in 2 variables.

First_Name = input("Enter First Name : ")
Last_Name  = input("\nEnter Last Name  : ")

print('-' *80)
print("\nReverse of First Name is : " + First_Name[::-1] + ' ' + " and last Name is : " + Last_Name[::-1])
print("\nReverse order : " + Last_Name  + "  " + First_Name)
print('-' *80)
```

```python
# Question 4

# Write a Python program to find the volume of a sphere with diameter 12 cm. Formula: V=4/3 * π * r 3

#Solution:
# Import math package to get value of PI
import math

# Calculate the radius of spehere from the given diameter formula is Radius = Diameter / 2
Diameter = int(input("Enter Diameter : "))
Radius = Diameter / 2
print('-' *80)
print ("\nRadius of the sphere : " + str(Radius))
print("\nVolume of sphere for a given " +  str(Diameter) + " is : " + str(4/3*math.pi*Radius**3))
print('-' *80)
```

```python
# Question 2 

#Create the below pattern using nested for loop in Python.
# *
# * *
# * * *
# * * * *
# * * * * *
# * * * *
# * * *
# * *
# *

Input_Pattern = input("Enter the pattern to be created : ")
Maxcount_Pattern = int(input("\nEnter Max count : "))

for Pattern in range (Maxcount_Pattern):
    for innerloop in range (Pattern):
        print(Input_Pattern + " " ,  end="")
    print('')
    
for Pattern in range (Maxcount_Pattern,0,-1):
    for innerloop in range (Pattern):
        print(Input_Pattern + " ",  end="")
    print('')    

```
